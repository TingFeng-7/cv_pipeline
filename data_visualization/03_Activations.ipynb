```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torch.autograd import grad

# 


# 
x = torch.linspace(-10, 10, 500, requires_grad=True)

#   
plt.figure(figsize=(10, 8)) 

# 
colors = ['b', 'g', 'r', 'c', 'm']

#   
for i, activation in enumerate(activations):
    plt.subplot(2, len(activations), i+1)
    y = activation(x)
    plt.plot(x.detach().numpy(), y.detach().numpy(), colors[i] , label=activation.__class__.__name__)
    # plt.plot(x.numpy(), activation(x).numpy(), colors[i])
    plt.title(activation.__class__.__name__)
    
#     
for i, activation in enumerate(activations):
    y = activation(x)
    dy_dx = grad(y.sum(), x, create_graph=True)[0] 
    plt.subplot(2, len(activations), i+len(activations)+1)
    plt.plot(x.detach().numpy(), dy_dx.detach().numpy(), colors[i], label=activation.__class__.__name__ + ' Derivative')
    plt.title(activation.__class__.__name__ + ' Derivative')
    
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
from torch.autograd import grad

# 
activations = [nn.ReLU(), nn.Sigmoid(), nn.Tanh(), nn.LeakyReLU(), nn.ELU()]  

# 
x = torch.linspace(-5, 5, 500, requires_grad=True)

# 
plt.subplot(2, 1, 1) 
for i, activation in enumerate(activations):
    y = activation(x)
    plt.plot(x.detach().numpy(), y.detach().numpy(), label=activation.__class__.__name__)
plt.legend()

#   
plt.subplot(2, 1, 2)
for i, activation in enumerate(activations):
    y = activation(x)
    dy_dx = grad(y.sum(), x, create_graph=True)[0] 
    plt.plot(x.detach().numpy(), dy_dx.detach().numpy(), label=activation.__class__.__name__ + ' Derivative')
plt.legend()

plt.show()
```
